```python
import pandas as pd
import io
import re
evol = pd.read_csv('evol_quedas_semnomes.csv.gz')
```

```python
pares = evol[['Unnamed: 0.1','REG. PACIENTE', 'date']].groupby(['REG. PACIENTE', 'date']).agg(['count'])
grouped = pd.DataFrame(pares['Unnamed: 0.1']['count'])

for par in grouped.index:
    reg, data = par
#     print(reg, data)
    
    evol_indexes = evol[(evol['REG. PACIENTE']==reg) & (evol['date']==data)].index

    print(evol_indexes)
    
    
    
    with open('{}_{}.txt'.format(reg, data), 'w') as queda:
        
        for idx in evol_indexes:
    
#         print(evol.loc[idx])
            
            reg_paciente = evol.loc[idx, 'REG. PACIENTE']
            data_evol = evol.loc[idx, 'DATA EVOL']
            resumo = str(evol.loc[idx, 'Resumo'])
            resumo = re.sub(u'[^a-zA-Z0-9áéíóúÁÉÍÓÚâêîôÂÊÎÔãõÃÕçÇ: ]', '', resumo)
            dados = str(evol.loc[idx, 'DADOS DA EVOLUÇÃO'])
            dados = re.sub(u'[^a-zA-Z0-9áéíóúÁÉÍÓÚâêîôÂÊÎÔãõÃÕçÇ: ]', '', dados)


            queda.write('Registro Paciente: {}.\n'.format(evol.loc[idx, 'Reg.Paciente']))
            queda.write('Data da Queda: {}.\n'.format(evol.loc[idx, 'Data Queda']))
            queda.write('Hora da Queda: {}.\n'.format(evol.loc[idx, 'Hora']))
            queda.write('Gênero: {}.\n'.format(evol.loc[idx, 'Gênero']))
            queda.write('Idade: {}.\n'.format(evol.loc[idx, 'Idade']))
            queda.write('UH: {}.\n'.format(evol.loc[idx, 'UH']))
            queda.write('Idade: {}.\n'.format(evol.loc[idx, 'Idade']))
            queda.write('Resumo: {}.\n'.format(resumo))
            queda.write('Houve dano ao paciente: {}.\n'.format(evol.loc[idx, 'Houve dano ao paciente?']))
            queda.write('Foi necessário algum procedimento: {}.\n'.format(evol.loc[idx, 'Foi necessário algum procedimento']))
            queda.write('Qual: {}.\n'.format(evol.loc[idx, 'Qual?']))
            queda.write('Condições do paciente antes da queda: {}.\n'.format(evol.loc[idx, 'Condições do paciente antes da queda']))
            queda.write('Risco do meio ambiente: {}.\n'.format(evol.loc[idx, 'Risco do meio ambiente']))
            queda.write('Local da queda: {}.\n'.format(evol.loc[idx, 'Local da queda']))
            queda.write('Outro setor.Qual: {}.\n'.format(evol.loc[idx, 'Outro setor:Qual?']))
            queda.write('Se foi na unidade de internação.Qual local: {}.\n'.format(evol.loc[idx, 'Se foi na unidade de internação.Qual local?']))
            queda.write('Fator desencadeante da queda: {}.\n'.format(evol.loc[idx, 'Fator desencadeante da queda']))
            queda.write('Registro Paciente: {}.\n'.format(reg_paciente))
            queda.write('Data Evolução: {}.\n'.format(data_evol))
            queda.write('Evolução: {}.\n'.format(dados))
        queda.close()
            # gera arquivos na pasta quedas com nomes {reg}_{date}.txt
    
    
```
